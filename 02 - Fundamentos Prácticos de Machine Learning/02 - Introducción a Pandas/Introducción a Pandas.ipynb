#### Importamos Pandas como pd

```python
import pandas as pd
```

#### Con pandas podemos crear series tal como la siguiente:

```python
series = pd.Series( [5, 10, 15, 20, 25, 30] )
series
```

```python
# Tipo de variable.
type(series)
```

#### Se puede elegir un valor dentro de la serie, recordando que la serie comienza con el índice cero.

```python
print(series[3])
print(series[5])
```

#### Podemos hacer series con palabras.

```python
# Serie con letras.
cadena = pd.Series( ['p', 'l', 'a', 't', 'z', 'i'] )
cadena
```

#### Pandas es bastante útil en la creación de Data Frames, veamos algunos ejemplos.

```python
# Convertamos en un data frame el siguiente conjunto de palabras.
lista = ['Hola', 'mundo', 'robótico']
df = pd.DataFrame(lista)
print(df)
```

```python
# Realicemos un data frame más complejo.
datos = { 'Nombre': ['Juan', 'Ana', 'Jose', 'Arturo' ], 
        'Edad' : [25, 18, 23, 27], 
        'Pais': ['MX', 'CO', 'BR', 'MX'] }
df = pd.DataFrame(datos)
df
```

```python
# Imprimamos las columnas del data frame.
print(df[[ 'Nombre', 'Edad', 'Pais' ]])
```
